```python
with open('benchmarkingResults.json', 'r') as f:
    data = json.load(f)
```

```python
request_per_second = [np.sum(benchmark) for benchmark in data.values()]
num_clients = [5,10,20,50,100]
```

```python
plt.plot(num_clients, request_per_second)
plt.xlabel('Num Clients')
plt.ylabel('Requests per Second')
plt.title('Requests per second vs. # of Clients')
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('lab3Results.json', 'r') as f:
    data = json.load(f)
```

```python
[10_000 / (x/1000) for x in data['5']]
```

```python
for key in data:
    curList = data[key]
    curList = [10_000/(x/1000) for x in curList]
    data[key] = curList
```

```python
request_per_second = [np.sum(benchmark) for benchmark in data.values()]
num_clients = [5,10,20,50,100]
```

```python
plt.plot(num_clients, request_per_second)
plt.xlabel('Num Clients')
plt.ylabel('Requests per Second')
plt.title('Requests per second vs. # of Clients')
```

```python
request_per_second = [np.mean(benchmark) for benchmark in data.values()]
num_clients = [5,10,20,50,100]

plt.plot(num_clients, request_per_second)
plt.xlabel('Num Clients')
plt.ylabel('Requests per Second per client')
plt.title('Requests per second vs. # of Clients')
```
